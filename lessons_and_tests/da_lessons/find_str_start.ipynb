```python
market_ticker = '^GSPC' #^GSPC
st = 'f'

stime='2016-01-01'
ftime=None
```

```python
def get_data_from_ticker(tick, stime, ftime, interval='1d'):
    ticker = yf.Ticker(tick)

    df = ticker.history(start=stime, end=ftime, interval=interval)
    x = pd.DataFrame(df)
    x.rename(columns={"Close": tick}, inplace=True)
    z = x.drop(columns=["Open", "High", "Low", "Dividends", "Stock Splits"])
    return z

df = pd.DataFrame(get_data_from_ticker(market_ticker, stime, ftime))



df.head(3)
```

```python
r = list(df['^GSPC'].values)
df['Signal'] = r

df['ma_10'] = df['^GSPC'].ewm(span=10, adjust=False).mean()
df['ma_20'] = df['^GSPC'].ewm(span=20, adjust=False).mean()
df['ma_50'] = df['^GSPC'].ewm(span=50, adjust=False).mean()
df['ma_100'] = df['^GSPC'].ewm(span=100, adjust=False).mean()
df['ma_150'] = df['^GSPC'].ewm(span=150, adjust=False).mean()
df['ma_200'] = df['^GSPC'].ewm(span=200, adjust=False).mean()


df
```
