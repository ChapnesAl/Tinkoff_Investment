upload data

```python
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df.info()
```

```python
df.head()
```

clearing the table

```python
''' убираем строки где нет данных '''

df = df.dropna()
print(df.shape)
```

```python
''' меняем некоторые типы данных object '''

df['User_Score'] = df.User_Score.astype('float64')
df['Year_of_Release'] = df.Year_of_Release.astype('int64')
df['User_Count'] = df.User_Count.astype('int64')
# df['Critic_Count'] = df.Critic_Count.astype('int64')
```

```python
''' оставляем колонки которыми мы будем пользоваться ''' 

useful_cols = ['Name', 'Platform', 'Year_of_Release', 'Genre', 
               'Global_Sales', 'Critic_Score', 'Critic_Count',
               'User_Score', 'User_Count', 'Rating'
              ]

df[useful_cols].head()
```

pandas PLOT tabple

```python
''' собираем таблицу из столбцов к которых есть "Sales" и год релиза'''
sales_df = df[[x for x in df.columns if 'Sales' in x] + ['Year_of_Release']]

''' группируем по году и получаем сумму продаж за год '''
group_1 = sales_df.groupby('Year_of_Release').sum()
group_1.head(3)
```

```python
''' строим граффик ''' 
group_1.plot()
```
